```python
import re
def extract_numbers(s):
    return [float(num) for num in re.findall(r'-?\d+\.?\d*', s)]

def parse_temperature_range(subtitle):
    numbers = extract_numbers(subtitle)
    if "or below" in subtitle:
        return {"min": float('-inf'), "max": numbers[0]}
    elif "or above" in subtitle:
        return {"min": numbers[0], "max": float('inf')}
    elif "to" in subtitle and len(numbers) == 2:
        return {"min": numbers[0], "max": numbers[1]}
```
